```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
```

```python
X=pd.read_csv('Linear_X_Train.csv')
Y=pd.read_csv('Linear_Y_Train.csv')
theta_list=np.load('Theta_list.npy')
```

```python
t0=theta_list[:,0]
t1=theta_list[:,1]
plt.ion()

for i in range(0,50,3):
    y_=t1[i]*X + t0[i]
    plt.scatter(X,Y)
    plt.plot(X,y_,'red')
    plt.draw()
    plt.pause(1)
    plt.clf()
     
    
```
